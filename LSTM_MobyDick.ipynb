```python
# Function to read a file and return it as a single string of that particular document
def read_file(file_path):
    with open(file_path) as f:
         str_text = f.read()
    return str_text
```

```python

# Load the text
file_path = "/content/moby_dick_e.txt"
text = read_file(file_path)

# Check the first few lines
print(text)
```

```python
import spacy

# Load SpaCy tokenizer (disable unnecessary components for speed)
nlp = spacy.load("en_core_web_sm", disable=["parser", "tagger", "ner"])
nlp.max_length = 1198623  # Adjust max length if needed
```

```python
def separate_punc(doc_text):
    return [token.text.lower() for token in nlp(doc_text) if token.text not in '\n\n \n\n\n!"-#$%&()--.*+,-/:;<=>?@[\\]^_`{|}~\t\n ']
```

```python

tokens = separate_punc(text)
```

```python
print("Sample Tokens:", tokens[:50])  # Preview first 50 tokens
```

```python
def clean_tokens(tokens):
    """
    Removes unwanted sections like Project Gutenberg text, chapter titles, and numbers.
    """
    # Find the start of Chapter 1 content
    try:
        start_index = tokens.index("loomings")  # Find where 'Loomings' appears
    except ValueError:
        print("❌ 'Loomings' not found. Check tokenization.")
        return []

    # Remove all content before 'Loomings' (start of the book)
    tokens = tokens[start_index:]

    # Remove "chapter" occurrences and numbers (e.g., "chapter 2", "3", etc.)
    tokens = [word for word in tokens if word not in ["chapter"] and not word.isdigit()]

    return tokens

# Apply the cleaning function
cleaned_tokens = clean_tokens(tokens)

print("✅ Cleaned Sample Tokens:", cleaned_tokens[:50])  # Preview first 50 cleaned tokens
```

```python
# Define sequence length (25 input + 1 target)
train_len = 25 + 1

# Create sequences
text_sequences = []

for i in range(train_len, len(cleaned_tokens)):
    seq = cleaned_tokens[i - train_len:i]  # Extract sequence
    text_sequences.append(seq)

# Print a sample sequence
print("✅ Sample Sequence:", text_sequences[0])
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
#from keras.preprocessing.text import Tokenizer

# Initialize tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(text_sequences)

# Convert text sequences into numerical sequences
sequences = tokenizer.texts_to_sequences(text_sequences)

# Check vocabulary size
vocabulary_size = len(tokenizer.word_counts) + 1  # +1 because indexing starts from 1
print("📌 Vocabulary Size:", vocabulary_size)

# Print a sample numerical sequence
print("✅ Sample Encoded Sequence:", sequences[0])
```

```python
# Check word frequencies
word_counts = tokenizer.word_counts

# Get vocabulary size (total unique words)
vocabulary_size = len(word_counts) + 1  # +1 because Keras indexing starts from 1

print("📌 Vocabulary Size:", vocabulary_size)
```

```python
import numpy as np

# Convert list of sequences to a NumPy array
sequences = np.array(sequences)

# Confirm conversion
print("✅ Data Type After Conversion:", type(sequences))
```

```python
from keras.utils import to_categorical

# First 24 words
print(sequences[:,:-1])

# last Word
print(sequences[:,-1])

X = sequences[:,:-1]
y = sequences[:,-1]
y = to_categorical(y, num_classes=vocabulary_size)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Input

def create_model(vocabulary_size, seq_len):
    model = Sequential()

    # Explicit Input layer
    model.add(Input(shape=(seq_len,)))

    # Embedding Layer (No input_length needed)
    model.add(Embedding(input_dim=vocabulary_size, output_dim=25))

    # LSTM layers
    model.add(LSTM(150, return_sequences=True))
    model.add(LSTM(150))

    # Fully connected layers
    model.add(Dense(150, activation='relu'))
    model.add(Dense(vocabulary_size, activation='softmax'))

    # Compile the model
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

    # Show model summary
    model.summary()

    return model

# Create the model
model = create_model(vocabulary_size, X.shape[1])
```

```python
model.fit(X, y, batch_size=128, epochs=300, verbose=1)
```

```python
import numpy as np
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
def generate_text(model, tokenizer, seq_len, seed_text, num_gen_words):
    """
    Generates text based on the trained model.

    Parameters:
    - model: Trained LSTM model.
    - tokenizer: Fitted Keras Tokenizer.
    - seq_len: Length of input sequences (25 in our case).
    - seed_text: Initial text to start generation.
    - num_gen_words: Number of words to generate.

    Returns:
    - Generated text.
    """

    output_text = []
    input_text = seed_text

    for _ in range(num_gen_words):
        # Convert input text to sequence
        encoded_text = tokenizer.texts_to_sequences([input_text])[0]

        # Pad the sequence to match training sequence length
        pad_encoded = pad_sequences([encoded_text], maxlen=seq_len, truncating='pre')

        # Predict next word
        predicted_probs = model.predict(pad_encoded, verbose=0)[0]  # Get probability distribution
        predicted_word_index = np.argmax(predicted_probs)  # Get index of most probable word

        # Retrieve word from tokenizer
        predicted_word = tokenizer.index_word.get(predicted_word_index, "")

        # Append predicted word
        input_text += " " + predicted_word
        output_text.append(predicted_word)

    return " ".join(output_text)
```
